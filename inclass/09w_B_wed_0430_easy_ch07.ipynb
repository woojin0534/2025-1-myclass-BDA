```python
p25, p50, p75 = mpg.hwy.quantile([.25,.5,.75])
```

```python
iqr = p75 - p25
iqr
```

```python
p75 + 1.5 * iqr  # 상한
```

```python
p25 - 1.5 * iqr  # 하한
```

```python
mpg.hwy.describe()
```

```python
mpg.hwy.describe()['50%']
```

```python
UF = p75 + 1.5 * iqr # 상한
```

```python
mpg[mpg.hwy > uf]
```

```python
mpg[mpg.hwy > uf][['hwy']]
```

```python
lf = p25 - 1.5 * iqr  # 하한
```

```python
mpg[mpg.hwy < lf]
```

```python
mpg.info()
```

```python
mpg.isna().sum()
```

```python
LF = p25 - 1.5 * iqr  # 하한
```

```python
mpg[ (mpg['hwy'] < 4.5) | (mpg['hwy'] > 40.5)]
```

```python
mpg[ (mpg['hwy'] < LF) | (mpg['hwy'] > UF)]
```

```python
mpg['hwy'] = np.where((mpg['hwy'] < LF) | (mpg['hwy'] > UF), np.nan, mpg['hwy'])
mpg.hwy.isna().sum()
```

```python
mpg[mpg.hwy.isna()]
```
