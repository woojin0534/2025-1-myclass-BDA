##mpg 데이터 활용 groupby() agg()

```python
import pandas as pd

mpg = pd.read_csv('data/mpg.csv')
mpg.info()
```

```python
mpg.sample(frac=.02)
```

```python
mpg.sample(frac=.01)
```

```python
# 제조 회사 및 구동 방식별 분리
# cty 평균 구하기
mpg.groupby(['manufacturer', 'drv']).agg(mean_cty = ('cty', 'mean')) # 시험 예상
```

```python
mpg.groupby(['drv','manufacturer']).agg(mean_cty = ('cty', 'mean'))
```

```python
mpg.groupby([ 'category','manufacturer']).agg(mean_cty = ('cty', 'mean'), mean_hwy = ('hwy', 'mean'))
```

```python
mpg.query('manufacturer == "audi"').groupby(['drv']).agg(n = ('drv', 'count'))
```

```python
mpg.query('manufacturer == "chevrolet"') \
   .groupby(['drv']) \
    .agg(n = ('drv', 'count'))
```

```python
mpg.query('manufacturer == "chevrolet"').groupby(['drv']).count()
```

```python
mpg.query('manufacturer == "chevrolet"').groupby(['drv']).count()[['model']]
```

```python
mpg.query('manufacturer == "chevrolet"').groupby(['drv']).count()[['model']].rename(columns = {'model' : 'n'})
```

```python
mpg.groupby('drv').agg(n = ('drv', 'count'))
```

```python
mpg[['drv', 'cyl']].value_counts()
```

```python
mpg.groupby(['drv', 'cyl']).agg('count')['model']
```

```python
mpg[['drv', 'cyl']].value_counts().to_frame()
```

```python
mpg[['drv', 'cyl']].value_counts().to_frame().query('count>40')
```

```python
mpg[['drv']].value_counts().to_frame()
```

```python
mpg[['drv']].value_counts().to_frame('n')
```

```python
mpg['drv'].value_counts().to_frame('n').rename_axis('drive')
```

```python
mpg.groupby('drv').agg(n = ('drv', 'count'))
```

```python
mpg.groupby('drv').agg(n = ('drv', 'count')).query('n > 100')
```

```python
mpg.query('category == "suv"') \
   .assign(total = (mpg['hwy'] + mpg['cty']) / 2) \
   .groupby('manufacturer') \
   .agg(mean_tot = ('total', 'mean')) \
   .sort_values('mean_tot', ascending = False) \
   .head(5)
```

## merge() 중간고사 예상 문

```python
import pandas as pd

df1 = pd.DataFrame([[1, 'apple'], [2, 'banana']],
                  columns = ['id', 'fruits'])
df1
```

```python
df2 = pd.DataFrame([[1, 3000], [5, 500]],
                  columns = ['id', 'price'])
df2
```

```python
pd.merge(df1, df2, on = 'id')
```

```python
pd.merge(df1, df2)
```

```python
pd.merge(df1, df2, how = 'inner')
```

```python
pd.merge(df1, df2,on = 'id',how='left')
```

```python
pd.merge(df1, df2,on = 'id',how='right')
```

```python
pd.merge(df1, df2 ,how='cross')
```

```python
df1.merge(df2, how='cross')
```
